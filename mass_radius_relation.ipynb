# Planet Mass and Radius Relation

```python
# Importing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
dataset = pd.read_csv('filtered_exoplanet_dataset_removed_outliers.csv')
```

```python
dataset.head(10)
```

#### Removing 5 exoplanets

```python
print("The shape of the dataset before removing 5 exoplanets is: ", dataset.shape)
```

```python
# We have 5,630 exoplanets in total, we must remove 5 exoplanets which orbit a white dwarf or brown dwarf star
# we should only have 5,635 exoplaents in the dataset which orbit a main sequence star (V), sub-giant (IV), or red giant (V)

# remove exoplanet NN Ser c
print("The exoplanet to drop is: ", dataset.iloc[4828, 0]) #.iloc[row number 4,828 where NN Ser c is present, first column]

# remove the row where the first column is NN Ser c
dataset = dataset[dataset.iloc[:,0] != 'NN Ser c']
print("The shape of the dataset after removing NN Ser c is: ", dataset.shape)
```

```python
# remove exoplanet NN Ser d
print("The exoplanet to drop is: ", dataset.iloc[4828, 0]) #.iloc[row number 4,828 where NN Ser d is present, first column]

# remove the row where the first column is NN Set d
dataset = dataset[dataset.iloc[:,0] != 'NN Ser d']
print("The shape of the dataset after removing NN Ser d is: ", dataset.shape)
```

```python
# remove exoplanet CFBDSIR J145829+101343 b
print("The exoplanet to drop is: ", dataset.iloc[83, 0]) #.iloc[row number 83 where CFBDSIR J145829+101343 b is present, first column]

# remove the row where the first column is CFBDSIR J145829+101343 b
dataset = dataset[dataset.iloc[:,0] != 'CFBDSIR J145829+101343 b']
print("The shape of the dataset after removing CFBDSIR J145829+101343 b is: ", dataset.shape)
```

```python
# remove exoplanet WISE J033605.05-014350.4 b
print("The exoplanet to drop is: ", dataset.iloc[5556, 0]) #.iloc[row number 5,556 where WISE J033605.05-014350.4 b is present, first column]

# remove the row where the first column is WISE J033605.05-014350.4 b
dataset = dataset[dataset.iloc[:,0] != 'WISE J033605.05-014350.4 b']
print("The shape of the dataset after removing WISE J033605.05-014350.4 b is: ", dataset.shape)
```

```python
# remove exoplanet WISEP J121756.91+162640.2 A b
print("The exoplanet to drop is: ", dataset.iloc[5556, 0]) #.iloc[row number 5,556 where WISEP J121756.91+162640.2 A b is present, first column]

# remove the row where the first column is WISEP J121756.91+162640.2 A b
dataset = dataset[dataset.iloc[:,0] != 'WISEP J121756.91+162640.2 A b']
print("The shape of the dataset after removing WISEP J121756.91+162640.2 A b is: ", dataset.shape)
```

#### Looking at the relation between mass and radius and creating planet classifications

```python
# Earth-Like exoplanet classification

# creating a condition to classify for earth-like exoplanets
earth_like_condition = (dataset['planet_rad_earth_rad'] <= 1.5) & (dataset['planet_mass_earth_mass'] >= 1) & (dataset['planet_mass_earth_mass'] < 2)

# count the number of exoplanets that satisfy this condition 
earth_like_count = dataset[earth_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are earth like is: ', earth_like_count)

# list the names of exoplanets that satisfy the condtion
earth_like_names = dataset[earth_like_condition]['planet_name']
print('The names of earth like exoplanets:')
print(earth_like_names)
```

```python
# Super-Earth-Like exoplanet classification
super_earth_like_condition = (dataset['planet_rad_earth_rad'] > 1.5) & (dataset['planet_rad_earth_rad'] <= 2) & (dataset['planet_mass_earth_mass'] >= 2) & (dataset['planet_mass_earth_mass'] < 8)

# count the number of exoplanets that satisfy this condition 
super_earth_like_count = dataset[super_earth_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are super earth like is: ', super_earth_like_count)

# list the names of exoplanets that satisfy the condtion
super_earth_like_names = dataset[super_earth_like_condition]['planet_name']
print('The names of super earth like exoplanets:')
print(super_earth_like_names)
```

```python
# Neptune-Like exoplanet classification
neptune_like_condition = (dataset['planet_rad_earth_rad'] > 2) & (dataset['planet_rad_earth_rad'] <= 7) & (dataset['planet_mass_earth_mass'] >= 8) & (dataset['planet_mass_earth_mass'] < 20)

# count the number of exoplanets that satisfy this condition 
neptune_like_count = dataset[neptune_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are neptune like is: ', neptune_like_count)

# list the names of exoplanets that satisfy the condtion
neptune_like_names = dataset[neptune_like_condition]['planet_name']
print('The names of neptune like exoplanets:')
print(neptune_like_names)
```

```python
# Super-Neptune-Like exoplanet classification
super_neptune_like_condition = (dataset['planet_rad_earth_rad'] >= 5) & (dataset['planet_rad_earth_rad'] <= 7) & (dataset['planet_mass_earth_mass'] >= 20) & (dataset['planet_mass_earth_mass'] <= 80)

# count the number of exoplanets that satisfy this condition 
super_neptune_like_count = dataset[super_neptune_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are super neptune like is: ', super_neptune_like_count)

# list the names of exoplanets that satisfy the condtion
super_neptune_like_names = dataset[super_neptune_like_condition]['planet_name']
print('The names of super neptune like exoplanets:')
print(super_neptune_like_names)
```

```python
# Jupiter-Like exoplanet classification
jupiter_like_condition = (dataset['planet_rad_earth_rad'] > 6) & (dataset['planet_rad_earth_rad'] <= 15) & (dataset['planet_mass_earth_mass'] >= 318) & (dataset['planet_mass_earth_mass'] <= 3180)

# count the number of exoplanets that satisfy this condition 
jupiter_like_count = dataset[jupiter_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are jupiter like is: ', jupiter_like_count)

# list the names of exoplanets that satisfy the condtion
jupiter_like_names = dataset[jupiter_like_condition]['planet_name']
print('The names of jupiter like exoplanets:')
print(jupiter_like_names)
```

```python
# Super-Jupiter-Like exoplanet classification
super_jupiter_like_condition = (dataset['planet_rad_earth_rad'] > 15) & (dataset['planet_mass_earth_mass'] > 3180)

# count the number of exoplanets that satisfy this condition 
super_jupiter_like_count = dataset[super_jupiter_like_condition].shape[0]

# print the number of exoplanets that satisfy the condition
print('The number of exoplanets that are super jupiter like is: ', super_jupiter_like_count)

# list the names of exoplanets that satisfy the condtion
super_jupiter_like_names = dataset[super_jupiter_like_condition]['planet_name']
print('The names of super jupiter like exoplanets:')
print(super_jupiter_like_names)
```

#### Making new dataframe to display results

```python
# Ensuring all panet classifications (Series) have the same length by filling with NaN

# max_length determines the maximum length between the different planet classifications (series)
max_length = max(earth_like_count, super_earth_like_count, neptune_like_count, super_neptune_like_count, jupiter_like_count, super_jupiter_like_count)
# creating the columns for the new dataframe
# .reindex extends the shorter dataframe to the maximum length, filling missing values with NaN.
column1 = dataset[earth_like_condition].reindex(range(max_length))
column2 = dataset[super_earth_like_condition].reindex(range(max_length))
column3 = dataset[neptune_like_condition].reindex(range(max_length))
column4 = dataset[super_neptune_like_condition].reindex(range(max_length))
column5 = dataset[jupiter_like_condition].reindex(range(max_length))
column6 = dataset[super_jupiter_like_condition].reindex(range(max_length))
```

```python
# creating new dataframe based on our results
new_dataframe = pd.DataFrame({
    'Earth Like': dataset[earth_like_condition]['planet_name'],
    'Super Earth Like': dataset[super_earth_like_condition]['planet_name'],
    'Neptune Like': dataset[neptune_like_condition]['planet_name'],
    'Super Neptune Like': dataset[super_neptune_like_condition]['planet_name'],
    'Jupiter Like': dataset[jupiter_like_condition]['planet_name'],
    'Super Jupiter Like': dataset[super_jupiter_like_condition]['planet_name']
})
```
